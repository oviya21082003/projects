# Converting text columns to numeric

```python
from sklearn.preprocessing import LabelEncoder
data3 = data2.copy()
data3 = data3.fillna(0)
ar = LabelEncoder()
pf = LabelEncoder()
bt = LabelEncoder()
st = LabelEncoder()
mz = LabelEncoder()

data3['AREA'] = ar.fit_transform(data3.AREA)
data3['PARK_FACIL'] = pf.fit_transform(data3.PARK_FACIL)
data3['BUILDTYPE'] = bt.fit_transform(data3.BUILDTYPE)
data3['STREET'] = st.fit_transform(data3.STREET)
data3['MZZONE'] = mz.fit_transform(data3.MZZONE)
data3.head(5)
```

# Correlation

```python
plt.figure(figsize=(10,10))
sns.heatmap(data3.corr(),annot=True,cbar=True,vmin=-1,vmax=1,square=False)
```

# Creating linear regression model

```python
from sklearn.preprocessing import StandardScaler as ss
```

```python
data3.head(5)
```

```python
sq_sc = ss()
tp_sc = ss()
```

```python
data3.INT_SQFT = sq_sc.fit_transform(data3[['INT_SQFT']])
data3.Total_price =tp_sc.fit_transform(data3[['Total_price']])
data3.head(5)
```

```python
relation = data3[['AREA', 'INT_SQFT', 'N_BEDROOM', 'N_BATHROOM', 'N_ROOM', 'PARK_FACIL',
       'BUILDTYPE', 'STREET', 'MZZONE', 'AGE_AT_SALE', 'Total_price']].corr()
relation
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data3[['AREA', 'INT_SQFT', 'N_BEDROOM', 'N_BATHROOM', 'N_ROOM', 'PARK_FACIL',
       'BUILDTYPE', 'STREET', 'MZZONE', 'AGE_AT_SALE', 'Total_price']]          #dependent data
y = data3[['Total_price']]
x.head(5),y.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.30, random_state=70)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# Training the model

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(X_train, y_train) 
```

```python
y_pred = model_lr.predict(X_test)
```

# Model Evaluation

```python
model_lr.coef_*100  
```

```python
model_lr.intercept_
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)   #RÂ² Score: Values closer to 1 indicate a better fit.
r2
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)   #error value
mse
```

```python
accuracy = 1-mse   #accuracy value
accuracy
```
