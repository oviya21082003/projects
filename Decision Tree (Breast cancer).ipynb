# Cost Complexity Pruning

Cost complexity pruning provides another option to control the size of a tree

```python
dtree.tree_.node_count
```

```python
dtree.get_n_leaves()
```

```python
prune_path=dtree.cost_complexity_pruning_path(X_train, y_train)
```

```python
alphas = prune_path.ccp_alphas   # ccp_alphas returns all the alpha values   8
alphas
```

```python
impurities = prune_path.impurities
impurities
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(alphas,impurities,marker='o',drawstyle="steps-post")
ax.set_xlabel('effective alphas')
ax.set_ylabel('Total impurities of leaves')
ax.set_title("Total impurity vs effective alphas for training set")
```

```python
new_dtrees = []
for alpha in alphas:
    new_dtree = DecisionTreeClassifier(criterion='gini',ccp_alpha=alpha)
    new_dtree.fit(X_train,y_train)
    new_dtrees.append(new_dtree)
```

```python
train_scores = [accuracy_score(y_train,new_dtree.predict(X_train)) for new_dtree in new_dtrees]
print(train_scores)
print('max train_score = {}'.format(max(train_scores)))
print('position of max score = {}'.format(np.argmax(train_scores)))
```

```python
test_scores = [accuracy_score(y_test,new_dtree.predict(X_test)) for new_dtree in new_dtrees]
print(test_scores)
print('max test_score = {}'.format(max(test_scores)))
print('position of max score = {}'.format(np.argmax(test_scores)))
```

```python
fig,ax = plt.subplots(figsize=(8,5))
ax.plot(alphas,train_scores,marker='o',label='train',drawstyle="steps-post")
ax.plot(alphas,test_scores,marker='o',label='test',drawstyle="steps-post")
ax.legend()
```
